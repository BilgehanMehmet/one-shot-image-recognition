```python
def plot(img):
    flag = False
    try:
        assert (img.shape == (105,105))
    except AssertionError as err:
        try:
            img = img.reshape(105,105)
        except:
            raise AssertionError("Image shape should be (105,105)")
    plt.imshow(img,cmap='gray')
    plt.show()
    

def val_acc(train_imgs,test_imgs,labels,sess,nn):
    val_acc = 0
#     for _ in range(n_trials // __N_TRIALS_PER_ITER):           
    for index in range(train_imgs.shape[0]): #20-way 20 trials. Each iteration is one one-shot trial.
        current_image_train_set = np.asarray(list(train_imgs[index]) * train_imgs.shape[0]).reshape(test_imgs.shape[0],105,105,1)
        current_train_img_label = labels[0,index]
        probs = sess.run(nn.probabilities, feed_dict= {nn.X:current_image_train_set , nn.X2:test_imgs})
        predicted_class = np.argmin(probs)
        if labels[0,predicted_class] == labels[0,index]:
            val_acc += 1
        print("probs:{}".format(probs))
        print("predicted class:{}".format(predicted_class))
        print("labels[predicted class]:{}, labels[true index]:{}".format(labels[0,predicted_class],labels[0,index]))
        print("val_acc:{}".format(val_acc))
        plot(train_imgs[predicted_class])
        plot(test_imgs[index])
#         x = input("")
    return val_acc/train_imgs.shape[0]
```

```python
data = Dataset.Loader(n_examples= 18,mode = "train_split", path = '/Users/mehmetbezcioglu/Documents/SEMESTER1/DeepLearning/one_shot_learning/image_similarity/datasets/images_background' , normalise = True)
```

```python
test_imgs, train_imgs, labels = data.generateOneShotTrials(data.X_validation,10)
```

```python
nn = neural_network.Siamese()
nn.logits = nn.forward_prop()
nn.probabilities = tf.nn.sigmoid(nn.logits)
saver = tf.train.Saver()
with tf.Session() as sess:
    saver.restore(sess,'/Users/mehmetbezcioglu/Documents/SEMESTER1/DeepLearning/one_shot_learning/image_similarity/models/pre_trained_weights/job_112/conv_siamese_model.ckpt')
    print("model restored")
#     sess.run(tf.global_variables_initializer())
    v = val_acc(train_imgs, test_imgs, labels,sess,nn)
#     sess.run()
print(v)
```
