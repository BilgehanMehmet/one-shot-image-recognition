## Pseudocode

-  as you iterate over each image, read idx1, and idx2.
-  use idx1 and idx2 to find the index of the string path in the .paths member of the class.
-  split the path.
-  edit the last 3rd item and re-concatenate the list as a new path.
   -  For 8 distortions e.g. english_distorted_1,english_distorted_2, ... , english_distorted_8
-  use the new path to save the augmented image.
